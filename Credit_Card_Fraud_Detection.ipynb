```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
import xgboost
from xgboost import XGBClassifier
```

```python
legit_data = data[data['Class'] == 0]
fraud_data = data[data['Class'] == 1]
```

```python
legit_sample = legit_data.sample(n=492)
```

```python
new_data = pd.concat([legit_sample,fraud_data],axis = 0)
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
x = new_data.iloc[:,:-1]
x.head()
```

```python
y = new_data.iloc[:,-1]
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,stratify = y,random_state=4)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(x_train,y_train)
```

```python
train_pred_1 = logistic_regression.predict(x_train)
test_pred_1 = logistic_regression.predict(x_test)
```

```python
Accuracy = {} 
```

```python
log_reg_train_accuracy = accuracy_score(train_pred_1,y_train)
log_reg_test_accuracy = accuracy_score(test_pred_1,y_test)
print("Training accuracy is ",log_reg_train_accuracy)
print("Testing accuracy is ",log_reg_test_accuracy)
Accuracy['Logistic_regression accuracy'] = log_reg_test_accuracy
```

```python
svm_model = SVC(kernel="linear")
svm_model.fit(x_train,y_train)
```

```python
train_pred_2 = svm_model.predict(x_train)
test_pred_2 = svm_model.predict(x_test)
```

```python
svm_train_accuracy = accuracy_score(train_pred_2,y_train)
svm_test_accuracy = accuracy_score(test_pred_2,y_test)
print("Training accuracy is ",svm_train_accuracy)
print("Testing accuracy is ",svm_test_accuracy)
Accuracy['SVM accuracy'] = svm_test_accuracy
```

```python
random_forest = RandomForestClassifier(n_estimators=10)
random_forest.fit(x_train,y_train)
```

```python
train_pred_3 = random_forest.predict(x_train)
test_pred_3 = random_forest.predict(x_test)
```

```python
rand_forest_train_accuracy = accuracy_score(train_pred_3,y_train)
rand_forest_test_accuracy = accuracy_score(test_pred_3,y_test)
print("Training accuracy is ",rand_forest_train_accuracy)
print("Testing accuracy is ",rand_forest_test_accuracy)
Accuracy['Random_Forest accuracy'] = rand_forest_test_accuracy
```

```python
bagging = BaggingClassifier(random_state = 100)
bagging.fit(x_train,y_train)
train_pred_4 = bagging.predict(x_train)
test_pred_4 = bagging.predict(x_test)
bagging_test_accuracy = accuracy_score(test_pred_4,y_test)
print("Testing accuracy is ",bagging_test_accuracy)
Accuracy['Bagging accuracy'] = bagging_test_accuracy
```

```python
xgboost = XGBClassifier()
xgboost.fit(x_train,y_train)
test_pred_5 = xgboost.predict(x_test)
xgb_accuracy = accuracy_score(test_pred_5,y_test)
print("Testing accuracy is ",xgb_accuracy)
Accuracy['XGBoost accuracy'] = xgb_accuracy
```

```python
Accuracy = pd.DataFrame([Accuracy]).T
Accuracy = Accuracy.rename(columns={0:'Accuracy_score'})
Accuracy
```
